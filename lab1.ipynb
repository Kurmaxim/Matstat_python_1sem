Раздел 8. Поведенческие факторы риска: табак, алкоголь, наркотические средства

Курите ли Вы в настоящее время

```python
df_norm = pd.read_excel("22_2022_СЗН.xlsx")
df = pd.DataFrame(df_norm)
df_norm
```

Нормальное распределение

```python
fig = plt.figure(figsize=(16, 8))
ax = sns.barplot(y=df_norm['Да, ежедневно '], x=df_norm["Возраст"])
plt.ylabel("Курящие ежедневно")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(df_norm['Да, ежедневно '], labels=df_norm["Возраст"], autopct='%1.1f%%', startangle=140)
plt.title('Процент курящих ежедневно по возрастным группам', fontsize=14)
plt.axis('equal')
plt.show()
```

Несмещенная медиана

```python
print(f"Среднее значение равно: {np.mean(df_norm['Да, ежедневно '])}")
print(f"Медиана равна: {statistics.median(df_norm['Да, ежедневно '])}")
if len(df_norm["Да, ежедневно "].unique()) == len(df_norm["Да, ежедневно "]):
    print("Мода не определена т.к. все значения уникальны")
else:
    print(f"Мода равна: {statistics.mode(df_norm['Да, ежедневно '])}")
print(f"Смещение медианы относительно среднего равно: {abs(np.mean(df_norm['Да, ежедневно ']) - statistics.median(df_norm['Да, ежедневно ']))}")
print(f"15-процентов размаха равны: {0.15 * (np.max(df_norm['Да, ежедневно ']) - np.min(df_norm['Да, ежедневно ']))}")
print("Медиана несмещенная")
```

Раздел 6. Занятия физкультурой и спортом

Наличие мест для занятий физической культурой и спортом

Равномерное распределение

```python
df_uniform = pd.DataFrame(pd.read_excel("21_2022_СЗН.xlsx"))
df_uniform = df_uniform.iloc[1:]
df_uniform
```

```python
fig = plt.figure(figsize=(16, 8))
ax = sns.barplot(y=df_uniform["Да, но я не пользуюсь ими"], x=df_uniform["Возраст"])
plt.show()
```

```python
mean_value = np.mean(df_uniform["Да, но я не пользуюсь ими"])
std_value = np.std(df_uniform["Да, но я не пользуюсь ими"])

plt.figure(figsize=(10, 5))
plt.bar(df_uniform["Возраст"], df_uniform["Да, но я не пользуюсь ими"])
plt.axhline(y=mean_value, color='r', linestyle='--', label=f'Среднее = {mean_value:.2f}')
plt.fill_between(df_uniform["Возраст"], mean_value - std_value, mean_value + std_value, color='r', alpha=0.2, label=f'±1 стандартное отклонение ({std_value:.2f})')
plt.xlabel('Возраст')
plt.ylabel('Процент')
plt.title('Распределение по категории "Да, но я не пользуюсь ими"')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(df_uniform["Да, но я не пользуюсь ими"], labels=df_uniform["Возраст"], autopct='%1.1f%%', startangle=140)
plt.title('Процент людей рядом с которыми присутствуют спортивные комплексы, но они ими не пользуются', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
print(f"Среднее значение равно: {np.mean(df_uniform['Да, но я не пользуюсь ими'])}")
print(f"Медиана равна: {statistics.median(df_uniform['Да, но я не пользуюсь ими'])}")
if len(df_uniform["Да, но я не пользуюсь ими"].unique()) == len(df_uniform["Да, но я не пользуюсь ими"]):
    print("Мода не определена т.к. все значения уникальны")
else:
    print(f"Мода равна: {statistics.mode(df_uniform['Да, но я не пользуюсь ими'])}")
print(f"Смещение медианы относительно среднего равно: {abs(np.mean(df_uniform['Да, но я не пользуюсь ими']) - statistics.median(df_uniform['Да, но я не пользуюсь ими']))}")
print(f"15-процентов размаха равны: {0.15 * (np.max(df_uniform['Да, но я не пользуюсь ими']) - np.min(df_uniform['Да, но я не пользуюсь ими']))}")
```

Раздел 2. Состояние здоровья населения

Оценка состояния здоровья в настоящее время

Смещенная медиана

```python
df = pd.DataFrame(pd.read_excel("12_2022_СЗН.xlsx"))
df
```

```python
fig = plt.figure(figsize=(16, 8))
ax = sns.barplot(y=df["Очень хорошее"], x=df["Возраст"])
plt.show()
```

```python
from scipy.stats import skew 
print(skew(df["Очень хорошее"], axis=0, bias=True))
print("Распределение имеет имеет положительную ассиметрию")
```

```python
plt.figure(figsize=(10, 10))
plt.pie(df["Очень хорошее"], labels=df["Возраст"], autopct='%1.1f%%', startangle=140)
plt.title('Процент людей состояние здоровья которых очень хорошее', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
print(f"Среднее значение равно: {np.mean(df['Очень хорошее'])}")
print(f"Медиана равна: {statistics.median(df['Очень хорошее'])}")
if len(df['Очень хорошее'].unique()) == len(df['Очень хорошее']):
    print("Мода не определена т.к. все значения уникальны")
else:
    print(f"Мода равна: {statistics.mode(df['Очень хорошее'])}")
print(f"Смещение медианы относительно среднего равно: {abs(np.mean(df['Очень хорошее']) - statistics.median(df['Очень хорошее']))}")
print(f"15-процентов размаха равны: {0.15 * (np.max(df['Очень хорошее']) - np.min(df['Очень хорошее']))}")
print("Медиана смещенна")
```

```python
data = pd.DataFrame(pd.read_csv("Car details v3.csv"))
data
```

```python
sns.relplot(x="year", y="selling_price", data=data, kind="scatter", col="transmission")
plt.show()
```

```python
sns.barplot(x="fuel", y="selling_price", data=data)
plt.show()
```

```python
data["engine"] = data["engine"].str.replace(' CC', '').astype(float)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="owner", y="engine", data=data)
plt.xticks(rotation=45)
plt.title('Диаграмма рассеивания для размера двигателя и количества владельцев')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["engine"], kde=True, bins=30)
plt.title('Распределение размера двигателя')
plt.xlabel('Engine (in CC)')
plt.show()
```

```python
data = data.dropna(subset=['engine'])
print(f"Стандартное отклонение равно: {data['engine'].std()}")
percentiles = np.percentile(data['engine'], [25, 50, 75])
print(f"25-й перцентиль равен: {percentiles[0]}")
print(f"50-й перцентиль равен: {percentiles[1]}")
print(f"75-й перцентиль равен: {percentiles[2]}")
print(f"Межквартильный интервал равен: {percentiles[2] - percentiles[0]}")
print(f"Размах равен: {data['engine'].max() - data['engine'].min()}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='owner', y='km_driven', data=data)
plt.xticks(rotation=45)
plt.title('Пробег машины в зависимости от количества прошлых владельцев')
plt.show()
```

```python
print(f"Стандартное отклонение равно: {data['km_driven'].std()}")
percentiles = np.percentile(data['km_driven'], [25, 50, 75])
print(f"25-й перцентиль равен: {percentiles[0]}")
print(f"50-й перцентиль равен: {percentiles[1]}")
print(f"75-й перцентиль равен: {percentiles[2]}")
print(f"Межквартильный интервал равен: {percentiles[2] - percentiles[0]}")
print(f"Размах равен: {data['km_driven'].max() - data['km_driven'].min()}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='owner', y='selling_price', data=data)
plt.xticks(rotation=45)
plt.title('Цена продажи в зависимости от количества прошлых владельцев')
plt.show()
```

```python
print(f"Стандартное отклонение равно: {data['selling_price'].std()}")
percentiles = np.percentile(data['selling_price'], [25, 50, 75])
print(f"25-й перцентиль равен: {percentiles[0]}")
print(f"50-й перцентиль равен: {percentiles[1]}")
print(f"75-й перцентиль равен: {percentiles[2]}")
print(f"Межквартильный интервал равен: {percentiles[2] - percentiles[0]}")
print(f"Размах равен: {data['selling_price'].max() - data['selling_price'].min()}")
```

Гипотезы:

Возраст машины: Машины с большим количеством предыдущих владельцев могут иметь более старый двигатель или более высокие пробеги, что может влиять на мощность двигателя.

Тип топлива: Дизельные двигатели могут иметь больше вариаций по мощности, что может привести к выбросам в данных.

Количество владельцев: Машины с большим количеством владельцев могут чаще поддаваться изменениям в техническом состоянии, что также может приводить к аномалиям в мощности двигателя.
